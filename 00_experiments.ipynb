```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset('tips')

# Create the histogram
ax = sns.histplot(data=data, x='total_bill', bins=10)

# Add labels to each bin
for patch in ax.patches:
    # Get the coordinates of the bar
    x = patch.get_x() + patch.get_width() / 2
    y = patch.get_height()
    
    # Add text annotation
    ax.text(x, y, f'{int(y)}', ha='center', va='bottom')

# Set labels and title
ax.set_xlabel('Total Bill')
ax.set_ylabel('Count')
ax.set_title('Histogram of Total Bill with Bin Labels')

plt.show()
```

```python
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
import scipy.stats as stats

# Sample data
data = px.data.tips()

# Extract the data for the histogram and KDE
x = data['total_bill']

# Create the histogram
hist = go.Histogram(
    x=x,
    nbinsx=20,
    histnorm='probability density',
    name='Histogram',
    opacity=0.75
)

# Create the KDE
kde_x = np.linspace(x.min(), x.max(), 1000)
kde_y = stats.gaussian_kde(x)(kde_x)

kde = go.Scatter(
    x=kde_x,
    y=kde_y,
    mode='lines',
    name='KDE',
    line=dict(width=2, color='red')
)

# Combine the histogram and KDE plot
fig = go.Figure(data=[hist, kde])

# Add labels and title
fig.update_layout(
    title='Histogram with KDE of Total Bill',
    xaxis_title='Total Bill',
    yaxis_title='Density',
    template='plotly_white'
)

# Show the plot
fig.show()
```
