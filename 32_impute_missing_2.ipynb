```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error, mean_absolute_percentage_error
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.drop('deck', axis=1, inplace=True)
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.info()
```

```python
df.head()
```

```python
columns = ['sex', 'embarked', 'who', 'class', 'embark_town', 'alive']

label_encoders = {}

for col in columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

    label_encoders[col] = le

label_encoders
```

```python
df
```

```python
df_with_missing = df[df['age'].isna()]

df_without_missing = df.dropna()
```

```python
df_with_missing.isnull().sum()
```

```python
df_with_missing.info()
```

```python
df_without_missing.info()
```

```python
df_with_missing.head()
```

```python
X1 = df_without_missing.drop('age', axis=1)
y1 = df_without_missing[['age']]

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2)
```

```python
rf_model = RandomForestRegressor(n_estimators=1000, random_state=42)
rf_model.fit(X1_train, y1_train)
```

```python
y1_pred = rf_model.predict(X1_test)

np.sqrt(mean_squared_error(y1_test, y1_pred)), r2_score(y1_test, y1_pred), mean_absolute_error(y1_test, y1_pred), mean_absolute_percentage_error(y1_test, y1_pred)
```

```python
y_pred = rf_model.predict(df_with_missing.drop('age', axis=1))
```

```python
df_with_missing['age'] = y_pred
```

```python
df_full = pd.concat((df_with_missing, df_without_missing), axis=0)
```
