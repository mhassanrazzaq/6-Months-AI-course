```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
df = sns.load_dataset('tips')
df
```

```python
# time as label encoding module
le = LabelEncoder()
ohe = OneHotEncoder()
oe = OrdinalEncoder()
```

```python
df['encoded_time'] = le.fit_transform(df['time'])
df
```

```python
df['encoded_time'].value_counts()
```

```python
# day to ordinal encoding by specific order
df['encoded_day'] = oe.fit_transform(df[['day']])
df
```

```python
print(df['day'].value_counts())
print('~~~~~~~~~~~~~~~~~~~~~')
print(df['encoded_day'].value_counts())
```

```python
oe2 = OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])

df['encoded_day2'] = oe2.fit_transform(df[['day']])
```

```python
df['encoded_day2'].value_counts().sort_values()
```

```python
ohe.fit_transform(df[['day']]).toarray()
```

```python
day_one_hot = ohe.fit_transform(df[['day']])

ohe2 = pd.DataFrame(day_one_hot.toarray(), columns=ohe.categories_[0])
ohe2
```

```python
df = pd.concat([df.reset_index(), ohe2.reset_index()], axis=1)
```
