```python
import pandas as pd

# Sample data
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]

# Using cut to create age groups
bins = [18, 25, 35, 60, 100]
labels = ['18-25', '26-35', '36-60', '61-100']
age_groups = pd.cut(ages, bins=bins, labels=labels, right=False)

print(age_groups)
```

```python
bins=[0,1,5,12,18,30,50,80]
labels=["Infants", "Toddlers", "Kids", "Teens", "Young", "Middle", "Old"]
df["bins"] = pd.cut(df.age, bins=bins, labels=labels)
```
