```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# data fetch from github

df = pd.read_csv("AirPassengers.csv")
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(df, y='#Passengers', x='Month')

plt.grid(True)
```

```python
# plot the same lineplot using plotly
plt.figure(figsize=(10,6))

fig = px.line(df, x='Month', y='#Passengers', title='Air Passengers')
fig.show()
```

```python
# decomposing the data into trend, seasonality and residuals
decomposed = seasonal_decompose(df['#Passengers'], model='multiplicative', period=12)
decomposed.plot()
```

```python
trend = decomposed.trend
seasonal = decomposed.seasonal
residual = decomposed.resid
```

```python
trend.plot()
```

```python
seasonal.plot()
```

```python
residual.plot()
```
