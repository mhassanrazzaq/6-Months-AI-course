```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(8,3))
sns.scatterplot(data=df, x="age", y='fare', hue='class', color="red")
```

```python
plt.figure(figsize=(8,3))
palette = {'First':"red", 'Second':"blue", "Third":'green'}
sns.scatterplot(data=df, x="age", y='fare', hue='class', palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
# plt.savefig('figure.jpeg', dpi=300, bbox_inches='tight')
# plt.savefig('figure.svg', dpi=300, bbox_inches='tight')
sns.set_style('darkgrid')
sns.set_theme()
```

```python
plt.figure(figsize=(10,3))
palette = {'First':"red", 'Second':"blue", "Third":'green'}
sns.boxplot(data=df, x="sex", y='age', hue='class')
# plt.xlabel('Class')
# plt.ylabel('Fare')
# plt.savefig('figure.jpeg', dpi=300, bbox_inches='tight')
# plt.savefig('figure.svg', dpi=300, bbox_inches='tight')
sns.set_style('darkgrid')
sns.set_theme()
plt.legend(loc='upper right', title='Class')
```

```python
plt.figure(figsize=(8,3))
palette = {'First':"red", 'Second':"blue", "Third":'green'}
sns.scatterplot(data=df[(df['class'] == 'First') & (df['survived'] == 1)], x="age", y='fare', hue='class', size='survived', palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
# plt.savefig('figure.jpeg', dpi=300, bbox_inches='tight')
# plt.savefig('figure.svg', dpi=300, bbox_inches='tight')
sns.set_style('darkgrid')
sns.set_theme()
```

```python
plt.figure(figsize=(8,3))
palette = {'First':"red", 'Second':"blue", "Third":'green'}
sns.scatterplot(data=df, x="age", y='fare', hue='class', size='age', sizes=(10, 100), palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
# plt.savefig('figure.jpeg', dpi=300, bbox_inches='tight')
# plt.savefig('figure.svg', dpi=300, bbox_inches='tight')
sns.set_style('darkgrid')
sns.set_theme()
```

```python
plt.figure(figsize=(10,3))
palette = {'First':"red", 'Second':"blue", "Third":'green'}
sns.boxenplot(data=df, x="sex", y='age', hue='class')
# plt.xlabel('Class')
# plt.ylabel('Fare')
# plt.savefig('figure.jpeg', dpi=300, bbox_inches='tight')
# plt.savefig('figure.svg', dpi=300, bbox_inches='tight')
sns.set_style('darkgrid')
sns.set_theme()
plt.legend(loc='upper right', title='Class')
# plt.xticks(labels=['MALE', 'FEMALE'])
# change the xticks labels
plt.xticks(ticks=[0, 1], labels=['MALE', 'FEMALE'])
```
