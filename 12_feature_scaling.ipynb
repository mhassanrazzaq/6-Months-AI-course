```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
data = {'Values' : [10,20,30,40,50]}
df = pd.DataFrame(data)
df
```

```python
scaler = MinMaxScaler()
df['scaled_values'] = scaler.fit_transform(df['Values'].values.reshape(-1,1))
df
```

```python
scaler = MinMaxScaler()
df['Scaled_values'] = scaler.fit_transform(df[['Values']])
df
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

df['Std Values'] = std_scaler.fit_transform(df[['Values']])
df
```

```python
from sklearn.preprocessing import RobustScaler

robust_scaler = RobustScaler()

df['Robust Values'] = robust_scaler.fit_transform(df[['Values']])
df
```

```python
# remove a column
df.drop('scaled_values', axis=1, inplace=True)
df
```

```python
# taking logrithmic scaler

df['Log Scaled'] = np.log(df['Values'])
df['Log10 Scaled'] = np.log10(df['Values'])
df
```

```python
data = {'Color':['Red', 'Green', 'Blue', 'Red']}
df = pd.DataFrame(data)
df

encoded_data = pd.get_dummies(df, columns=['Color'])
encoded_data
```

```python
from sklearn.preprocessing import LabelEncoder

data2 = {'Animal' : ['Dog', 'Cat', 'Bird', 'Dog']}
df2 = pd.DataFrame(data2)
df2

label_encoder = LabelEncoder()
df2['Labeled'] = label_encoder.fit_transform(df2['Animal'])
df2
```

```python
from sklearn.preprocessing import OrdinalEncoder

data3 = {'Animal' : ['Dog', 'Cat', 'Bird', 'Dog']}
df3 = pd.DataFrame(data3)
df3

Ordinal_encoder = OrdinalEncoder()
df3['Labeled'] = Ordinal_encoder.fit_transform(df3[['Animal']])
df3
```
